```python
'''
Khởi động 1: Viết chương trình in ra tất cả các số chia hết cho cả 5 và 7 trong khoảng từ 1500 đến 2700
'''
so_1 = 5
so_2 = 7

so = 1500
while so <= 2700:
    if so % (so_1 * so_2) == 0:
        print(so)

    so += 1
```
