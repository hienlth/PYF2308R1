## Bài 1: reversed list
Hãy viết một chương trình để đảo ngược các phần tử trong list

```python
aList = [100, 200, 300, 400, 500]
```

**Output:**

```python
[500, 400, 300, 200, 100]
```

```python
aList = [100, 200, 300, 400, 500]
# code below this line
for index in range(len(aList) - 1, -1, -1):
    print(aList[index], end=", ")
# code above this line
```

```python
# Tung example
aList = [100, 300, 250, 400, 500]
# code below this line
aList.reverse()
print(aList)
```

```python
# Hoc example
a = [100, 200, 300, 400, 500]
b = a[::-1]
print(b)
```

## Bài 2: concatenation
Hãy tạo ra một list mới bằng cách cộng các phần tử tương ứng của list cũ theo index tương ứng

```python
list1 = ["M", "na", "i", "Ke"]
list2 = ["y", "me", "s", "lly"]
```

**Output:**

```python
['My', 'name', 'is', 'Kelly']
```

```python
list1 = ["M", "na", "i", "Ke"]
list2 = ["y", "me", "s", "lly"]
# code below this line
result = []
for i in range(0, len(list1)):
    # result.append(list1[i] + list2[i])
    result.insert(i, list1[i] + list2[i])
print(result)

# code above this line
```

## Bài 3: squared list
Cho sẵn một list chứa các số tự nhiên. Bình phương các phần tử trong list và lưu vào một list mới.

```python
aList = [1, 2, 3, 4, 5, 6, 7]
```

**Output**

```python
# code below this line
aList = [1, 2, 3, 4, 5, 6, 7]
result = []
for item in aList:
    result.append(item ** 2)

print(result)
# code above this line
```

```python
aList = [1, 2, 3, 4, 5, 6, 7]
result = list(map(lambda x: x ** 2, aList))
print(result)
```

```python
aList = [1, 2, 3, 4, 5, 6, 7]
result = [i ** 2 for i in aList]
print(result)
```

## Bài 4: concatenation 2
Tạo ra một list mới từ 2 list ban đầu theo thứ tự như sau:

```python
list1 = ["Hello ", "take "]
list2 = ["Dear", "Sir"]
```

**Output**

```python
['Hello Dear', 'Hello Sir', 'take Dear', 'take Sir']
```

```python
list1 = ["Hello ", "take "]
list2 = ["Dear", "Sir"]
# code below this line
list3 = []
for a in list1:
    for b in list2:
        list3.append(a + b)
print(list3)
# code above this line
```

## Bài 5: print list
Cho 2 list có cùng số phần tử với nhau. Hãy in các thành phần trong 2 list theo format dưới đây:

```python
list1 = [10, 20, 30, 40]
list2 = [100, 200, 300, 400]
```

**Output**

```python
10 400
20 300
30 200
40 100
```

```python
list1 = [10, 20, 30, 40]
list2 = [100, 200, 300, 400]
# code below this line
N = len(list1)
for index in range(N):
    print(f"{list1[index]} {list2[- index - 1]}") #list2 xài chỉ số âm
    print(f"{list1[index]} {list2[N - 1 - index]}") #list2 xài chỉ số dương
# code above this line
```

```python
# Bài giảng của Ninh Trần
list1 = [10, 20, 30, 40]
list2 = [100, 200, 300, 400]
list2.reverse()
for item1, item2 in zip(list1, list2):
    print(f"{item1} - {item2}")
```

## Bài 6: remove empty
Hãy xóa tất cả empty string trong list đã cho.

```python
list1 = ["Mike", "", "Emma", "Kelly", "", "Brad"]
```

**Output**

```python
["Mike", "Emma", "Kelly", "Brad"]
```

```python
list1 = ["Mike", "", "Emma", "Kelly", "", "Brad", "", ""]
# code below this line
while "" in list1:
    list1.remove("")
print(list1)
# code above this line
```

## Bài 7: list replacement
Hãy tìm và thay thế số 20 **đầu tiên** trong list thành 200.

```python
list1 = [5, 10, 15, 20, 25, 50, 20]
```

**Output**

```python
[5, 10, 15, 200, 25, 50, 20]
```

```python
list1 = [5, 10, 15, 20, 25, 50, 20]
# code below this line
for i in range(len(list1)):
    if list1[i] == 20:
        list1[i] = 200
        break

print(list1)
# code above this line
```

## Bài 8: list removement
Hãy xóa bỏ **tất cả** số 20 ra khỏi list

```python
list1 = [5, 20, 15, 20, 25, 50, 20]
```

**Output:**

```python
[5, 15, 25, 50]
```

```python
list1 = [5, 20, 15, 20, 25, 50, 20]
# code below this line
while 20 in list1:
    list1.remove(20)

print(list1)   
# code above this line
```
