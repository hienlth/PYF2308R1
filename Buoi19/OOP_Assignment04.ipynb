```python
"""
Ví dụ mô phỏng sự hoạt động của một chiếc đèn pin.
Trong ví dụ này ta có 2 lớp đối tượng cơ bản là Đèn (FlashLamp) và Pin (Battery).
Đối tượng pin (Battery) mang trong mình thông tin về trạng thái năng lượng của nó (energy), 
đối tượng đèn sẽ sử dụng một đối tượng pin để cung cấp năng lượng cho hoạt động chiếu sáng,
khi đó ta nói có sự tương tác và trao đổi thông tin giữa hai đối tượng đèn và pin.
"""
class Battery:
    def __init__(self):
        self.energy = 10
    
    def set_energy(self, x_energy):
        self.energy = x_energy
        
    def get_energy(self):
        return self.energy
    
    def decrease_energy(self):
        if self.energy > 0:
            self.energy -= 2
        
        
class FlashLamp:
    def __init__(self):
        self.status = False
        self.battery = None
        
    def set_battery(self, b):
        self.battery = b
    
    def get_battery_info(self):
        return self.battery.get_energy()
    
    def turn_on(self):
        self.status = True
        print("Turn on")
        self.battery.decrease_energy()
        if self.get_battery_info() <= 0:
            print("Light off")
            self.status = False
        else:
            print("Light on")
        
    def turn_off(self):
        self.status = False
        print("Turn off")
        
        
class TestFlashLamp:
    def main(self):
        battery = Battery()
        flash_lamp = FlashLamp()
        flash_lamp.set_battery(battery)
        
        for i in range(10):
            print(f"i = {i}")
            flash_lamp.turn_on()
            flash_lamp.turn_off()
            print("------------------------------")
            
        print(f"Remaining energy in flashmlamp battery: {flash_lamp.get_battery_info()}")
        
        
TestFlashLamp().main()
```
